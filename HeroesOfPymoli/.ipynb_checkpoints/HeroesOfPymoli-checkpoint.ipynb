### Initial Setup and Reading of Data File
* Instructions have been included for each segment. You do not have to follow them exactly, but they are included to help you think through the steps.

```python
# Dependencies and Setup
import pandas as pd

# Set input file containing original purchasing data
RawPurchaseDataFile = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
RawPurchasesDF = pd.read_csv(RawPurchaseDataFile)

# Check for missing data
RawPurchasesDF.count()
# All columns have the same amount of data: Data cleaning not needed
```

## Player Count

* Display the total number of players


```python
# Create and display data frame for total number of players
TotalPlayers = RawPurchasesDF["SN"].nunique()
TotalPlayersDF = pd.DataFrame({"Total Players": [TotalPlayers]})
TotalPlayersDF
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Determine: number of unique items, average purchase price,
#            number of purchases, and total revenue
NumUniqueItems = RawPurchasesDF["Item ID"].nunique()
AvgPrice = RawPurchasesDF["Price"].mean()
NumPurchases = RawPurchasesDF["Purchase ID"].count()
TotalRevenue = RawPurchasesDF["Price"].sum()

# Create and display data frame for results of above analysis
PurchaseSummaryDF = pd.DataFrame({
    "Number of Unique Items": [NumUniqueItems],
    "Average Price": f"${AvgPrice:.2f}",
    "Number of Purchases": NumPurchases,
    "Total Revenue": f"${TotalRevenue:,.2f}"
    })
PurchaseSummaryDF
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Determine the number and percentage of unique players for each gender specification
TotalMalePlayers = RawPurchasesDF[RawPurchasesDF["Gender"]=="Male"].loc[:,"SN"].nunique()
TotalFemalePlayers = RawPurchasesDF[RawPurchasesDF["Gender"]=="Female"].loc[:,"SN"].nunique()
TotalOtherPlayers = RawPurchasesDF[RawPurchasesDF["Gender"]=="Other / Non-Disclosed"].loc[:,"SN"].nunique()
PercentMalePlayers = TotalMalePlayers / TotalPlayers  # TotalPlayers was calculated in a previous cell
PercentFemalePlayers = TotalFemalePlayers / TotalPlayers
PercentOtherPlayers = TotalOtherPlayers / TotalPlayers

# Create and display data frame for results of above analysis with proper output formatting
PlayerGenderStatsDF = pd.DataFrame(
    {
    "": ["Male", "Female", "Other / Non-Disclosed"],
    "Total Count": [TotalMalePlayers, TotalFemalePlayers, TotalOtherPlayers],
    "Percentage of Players": [f"{PercentMalePlayers:.2%}", f"{PercentFemalePlayers:.2%}", f"{PercentOtherPlayers:.2%}"]
    })
PlayerGenderStatsDF.set_index("")
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculate how many purchases were made by each gender
GenderPurchaseCounts = RawPurchasesDF.groupby(["Gender"]).count()["Purchase ID"].rename("Purchase Count")

# Calculate avg. purchase price for each gender
GenderPurchaseAvg = RawPurchasesDF.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")

# Calculate the total spent for each gender
GenderPurchaseTotals = RawPurchasesDF.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

# Create data frame for purchasing statistics by gender
PurchaseStatsByGenderDF = pd.DataFrame({
    "Purchase Count": GenderPurchaseCounts,
    "Average Purchase Price": GenderPurchaseAvg,
    "Total Purchase Value": GenderPurchaseTotals
})

# Create another column for the data frame that shows the avg. purchase per person for each gender
PurchaseStatsByGenderDF["Avg Total Purchase per Person"] = [
    PurchaseStatsByGenderDF.loc["Female", "Total Purchase Value"] / TotalFemalePlayers,  # TotalFemalePlayer calculated in a previous cell
    PurchaseStatsByGenderDF.loc["Male", "Total Purchase Value"] / TotalMalePlayers,  #TotalMalePlayers calculated in a previous cell
    PurchaseStatsByGenderDF.loc["Other / Non-Disclosed", "Total Purchase Value"] / TotalOtherPlayers  #TotalOtherPlayers calculated in a previous cell]
    ]

# Format the values in the columns and display the data frame
PurchaseStatsByGenderDF["Average Purchase Price"] = PurchaseStatsByGenderDF["Average Purchase Price"].map('${:,.2f}'.format)
PurchaseStatsByGenderDF["Total Purchase Value"] = PurchaseStatsByGenderDF["Total Purchase Value"].map('${:,.2f}'.format)
PurchaseStatsByGenderDF["Avg Total Purchase per Person"] = PurchaseStatsByGenderDF["Avg Total Purchase per Person"].map('${:,.2f}'.format)
PurchaseStatsByGenderDF
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish age division bins
AgeDivisions = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
AgeGroups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Add a new column to the raw purchases data frame that contains the the Age Group for each purchase
RawPurchasesDF["Age Group"] = pd.cut(RawPurchasesDF["Age"], AgeDivisions, labels=AgeGroups, include_lowest=True)

# Create a smaller data frame showing just players and their age groups with duplicate players removed
PlayersAgeGroupDF = RawPurchasesDF[["SN", "Age Group"]].drop_duplicates()

# Determine the number of players in each age group
AgeGroupPlayerCounts = PlayersAgeGroupDF["Age Group"].value_counts()

# Calculate what percent of all players were present in each age group
AgeGroupRepresentation = AgeGroupPlayerCounts / TotalPlayers  # TotalPlayers was calculated in a previous cell

# Create data frame for player age statistics
PlayerAgeStatsDF = pd.DataFrame({
    "Total Count": AgeGroupPlayerCounts,
    "Percentage of Players": AgeGroupRepresentation
})

# Format and display sorted data frame
PlayerAgeStatsDF["Percentage of Players"] = PlayerAgeStatsDF["Percentage of Players"].map("{:.2%}".format)
PlayerAgeStatsDF.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Determine number of purchases in each age group
AgeGroupPurchaseCounts = RawPurchasesDF.groupby(["Age Group"]).count()["Purchase ID"].rename("Purchase Count")

# Calculate avg. purchase price in each age group
AgeGroupPurchaseAvg = RawPurchasesDF.groupby(["Age Group"]).mean()["Price"].rename("Average Purchase Price")

# Calculate total purchases per age group
AgeGroupPurchaseTotals = RawPurchasesDF.groupby(["Age Group"]).sum()["Price"].rename("Total Purchase Value")

# Calculate avg. purchase per person in each age group
AvgPurchasePerPersonInAgeGroup = AgeGroupPurchaseTotals / AgeGroupPlayerCounts  # AgeGroupPlayerCounts calculated in a previous cell

# Create data frame for purchase statistics by age group
PurchaseStatsByAgeGroupDF = pd.DataFrame({
    "Age Ranges": AgeGroups,  # AgeGroups defined in a previous cell
    "Purchase Count": AgeGroupPurchaseCounts,
    "Average Purchase Price": AgeGroupPurchaseAvg,
    "Total Purchase Value": AgeGroupPurchaseTotals,
    "Avg Total Purchase per Person": AvgPurchasePerPersonInAgeGroup
})

# Format and display sorted data frame
PurchaseStatsByAgeGroupDF["Average Purchase Price"] = PurchaseStatsByAgeGroupDF["Average Purchase Price"].map('${:,.2f}'.format)
PurchaseStatsByAgeGroupDF["Total Purchase Value"] = PurchaseStatsByAgeGroupDF["Total Purchase Value"].map('${:,.2f}'.format)
PurchaseStatsByAgeGroupDF["Avg Total Purchase per Person"] = PurchaseStatsByAgeGroupDF["Avg Total Purchase per Person"].map('${:,.2f}'.format)
PurchaseStatsByAgeGroupDF.sort_index().set_index("Age Ranges")
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Determine the number of purchases by each player
PlayerPurchaseCounts = RawPurchasesDF.groupby(["SN"]).count()["Purchase ID"].rename("Purchase Count")

# Calculate the average purchase price for each player
PlayerPurchaseAvg = RawPurchasesDF.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")

# Sum the total purchases for each player
PlayerPurchaseTotals = RawPurchasesDF.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

# Create data frame for each player's purchasing statistics
PlayerSpendingStatsDF = pd.DataFrame({
    "Purchase Count": PlayerPurchaseCounts,
    "Average Purchase Price": PlayerPurchaseAvg,
    "Total Purchase Value": PlayerPurchaseTotals
})

# Sort, format, and display data frame
PlayerSpendingStatsDF = PlayerSpendingStatsDF.sort_values("Total Purchase Value", ascending=False)
PlayerSpendingStatsDF["Average Purchase Price"] = PlayerSpendingStatsDF["Average Purchase Price"].map('${:,.2f}'.format)
PlayerSpendingStatsDF["Total Purchase Value"] = PlayerSpendingStatsDF["Total Purchase Value"].map('${:,.2f}'.format)
PlayerSpendingStatsDF.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a smaller data frame containing only the columns for Item ID, Item Name, and Price
ItemsPurchasedDF = RawPurchasesDF[["Item ID", "Item Name", "Price"]]

# Count the number of purchases each item
ItemPurchaseCounts = ItemsPurchasedDF.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

# Calculate the average price for each item
ItemPurchaseAvg = ItemsPurchasedDF.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")

# Sum the total spent for each item
ItemPurchaseTotals = ItemsPurchasedDF.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

# Create data frame to each item's sales
PurchaseStatsByItemDF = pd.DataFrame({
    "Purchase Count": ItemPurchaseCounts,
    "Item Price": ItemPurchaseAvg,
    "Total Purchase Value": ItemPurchaseTotals
})

# Sort, format, and display data frame
MostPopularItemDF = PurchaseStatsByItemDF.sort_values("Purchase Count", ascending=False)
MostPopularItemDF["Item Price"] = MostPopularItemDF["Item Price"].map('${:,.2f}'.format)
MostPopularItemDF["Total Purchase Value"] = MostPopularItemDF["Total Purchase Value"].map('${:,.2f}'.format)
MostPopularItemDF.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort PurchaseStatsByItemDF by Total Purchase Value
MostProfitableItemDF = PurchaseStatsByItemDF.sort_values("Total Purchase Value", ascending=False)

# Format and display data frame
MostProfitableItemDF["Item Price"] = MostProfitableItemDF["Item Price"].map('${:,.2f}'.format)
MostProfitableItemDF["Total Purchase Value"] = MostProfitableItemDF["Total Purchase Value"].map('${:,.2f}'.format)
MostProfitableItemDF.head()
```
